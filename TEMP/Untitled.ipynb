```python
# url = 'http://www.perthmint.com.au/treasury/gold-Current.csv?_ga=2.217144726.1877785870.1557661316-1377825704.1557661316'

### OLD DATA

url = 'http://www.perthmint.com.au/treasury/gold.csv?_ga=2.259620106.1877785870.1557661316-1377825704.1557661316'

df = pd.read_csv(url,header=0)
df = df[['Market','Perth Mint Spot.15']]
df.columns = ['Date','AskPrice_Avg_AUD']
df.dropna(axis=0,how='any',inplace=True)
df = df.iloc[4:,:]
df.Date = pd.to_datetime(df.Date)

df.AskPrice_Avg_AUD = df.AskPrice_Avg_AUD.str.strip()
df.AskPrice_Avg_AUD = df.AskPrice_Avg_AUD.str.replace(',', '')
df.AskPrice_Avg_AUD = df.AskPrice_Avg_AUD.astype('float')
# df.set_index('Date',drop=True,inplace=True)

df.to_json('gold_askprice_avg_aud.json',orient='records')
```

```python
def data_update_GOLD(url = 'http://www.perthmint.com.au/treasury/gold-Current.csv'):

    filepath = "../media_files/datasets/gold_askprice_avg_aud.json"
    old_data = pd.read_json(filepath,orient='records')

    df = pd.read_csv(url,header=0)
    df = df[['Market','Perth Mint Spot.15']]
    df.columns = ['Date','AskPrice_Avg_AUD']
    df.dropna(axis=0,how='any',inplace=True)
    df = df.iloc[4:,:]
    df.Date = pd.to_datetime(df.Date)

    df.AskPrice_Avg_AUD = df.AskPrice_Avg_AUD.str.strip()
    df.AskPrice_Avg_AUD = df.AskPrice_Avg_AUD.str.replace(',', '')
    df.AskPrice_Avg_AUD = df.AskPrice_Avg_AUD.astype('float')
    old_data = old_data.append(df)

    old_data.to_json(filepath,orient='records')

    old_data.drop_duplicates(keep='last',inplace=True)
```
